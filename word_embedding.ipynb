```python
# Import necessary libraries
import numpy as np
import matplotlib.pyplot as plt
from sklearn.manifold import TSNE

# Step 1: Download and Load GloVe Embeddings
!wget http://nlp.stanford.edu/data/glove.6B.zip
!unzip glove.6B.zip -d glove.6B

# Define the path to the 50D GloVe file
glove_file_path = "glove.6B/glove.6B.50d.txt"

# Load GloVe embeddings
def load_glove_embeddings(file_path):
    embeddings = {}
    with open(file_path, "r", encoding="utf-8") as f:
        for line in f:
            values = line.split()
            word = values[0]
            vector = np.array(values[1:], dtype="float32")
            embeddings[word] = vector
    return embeddings

print("Loading GloVe embeddings...")
embeddings = load_glove_embeddings(glove_file_path)
print(f"Loaded {len(embeddings)} word vectors.")

# Step 2: Select a Subset of Words to Visualize
# We'll use common English words for this demonstration
selected_words = list(embeddings.keys())[:100]  # Select the first 500 words
word_vectors = np.array([embeddings[word] for word in selected_words])

# Step 3: Apply t-SNE for Dimensionality Reduction
print("Applying t-SNE...")
tsne_model = TSNE(n_components=2, random_state=42, perplexity=30, n_iter=1000)
reduced_vectors = tsne_model.fit_transform(word_vectors)

# Step 4: Plot the Reduced Embeddings
def plot_embeddings_2d(vectors, labels):
    plt.figure(figsize=(16, 12))
    plt.scatter(vectors[:, 0], vectors[:, 1], s=10, alpha=0.7)
    for i, label in enumerate(labels):
        plt.annotate(label, (vectors[i, 0], vectors[i, 1]), fontsize=9, alpha=0.8)
    plt.title("t-SNE Visualization of GloVe Word Embeddings")
    plt.show()

print("Plotting...")
plot_embeddings_2d(reduced_vectors, selected_words)
```

```python
import gensim.downloader as api
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

# Load a pre-trained word embedding model (e.g., Word2Vec)
model = api.load("word2vec-google-news-300")  # You can also try "word2vec-google-news-300"

# Words to visualize
words = ['happy', 'joyful', 'cat', 'dog','king','queen','zebra','rat','sad','worried']

# Extract embeddings for the words
embeddings = [model[word] for word in words]

# Apply PCA to reduce dimensions to 2D
pca = PCA(n_components=2)
reduced_embeddings = pca.fit_transform(embeddings)

# Visualization
plt.figure(figsize=(8, 6))
for word, coord in zip(words, reduced_embeddings):
    plt.scatter(coord[0], coord[1], label=word)
    plt.text(coord[0] + 0.01, coord[1] + 0.01, word, fontsize=12)

plt.title("Word Embedding Visualization (2D PCA)")
plt.xlabel("PCA Dimension 1")
plt.ylabel("PCA Dimension 2")
plt.legend()
plt.grid(True)
plt.show()
```

```python
from gensim.models import KeyedVectors
import numpy as np

# Load pre-trained Word2Vec model (Google News vectors)
print("Loading Word2Vec model...")
word2vec_model = KeyedVectors.load_word2vec_format('GoogleNews-vectors-negative300.bin', binary=True)

def cosine_similarity(vec1, vec2):
    """Compute cosine similarity between two vectors."""
    return np.dot(vec1, vec2) / (np.linalg.norm(vec1) * np.linalg.norm(vec2))

def euclidean_distance(vec1, vec2):
    """Compute Euclidean distance between two vectors."""
    return np.linalg.norm(vec1 - vec2)

# Word pairs
pairs = [("cat", "dog"), ("happy", "joyful")]

for word1, word2 in pairs:
    vec1 = word2vec_model[word1]
    vec2 = word2vec_model[word2]

    # Calculate distances
    cosine_sim = cosine_similarity(vec1, vec2)
    euclidean_dist = euclidean_distance(vec1, vec2)

    print(f"Words: ({word1}, {word2})")
    print(f"Cosine Similarity: {cosine_sim:.4f}")
    print(f"Euclidean Distance: {euclidean_dist:.4f}\n")
```
