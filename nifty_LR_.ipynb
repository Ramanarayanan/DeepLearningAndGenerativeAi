```python
import yfinance as yf
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import datetime
from sklearn.linear_model import LinearRegression

def fetch_nifty_data():
    # Use the Nifty 50 index ticker symbol for Nifty 50 (NSE)
    nifty_ticker = "^NSEI"  # Nifty 50 index on the National Stock Exchange (NSE)

    # Calculate the start date for 20 years ago
    end_date = datetime.datetime.today().strftime('%Y-%m-%d')
    start_date = (datetime.datetime.today() - datetime.timedelta(days=20*365)).strftime('%Y-%m-%d')

    # Fetch historical data for the last 20 years using yfinance
    nifty_data = yf.download(nifty_ticker, start=start_date, end=end_date)
    print(f"Nifty 50 Historical Data (From {start_date} to {end_date}):")
    print(nifty_data.head())

    return nifty_data

def apply_linear_regression(nifty_data):
    # Prepare the data for linear regression
    nifty_data['Date'] = pd.to_datetime(nifty_data.index)
    nifty_data['Date_Ordinal'] = nifty_data['Date'].apply(lambda date: date.toordinal())  # Convert date to ordinal

    # Prepare the features (Date_Ordinal) and target (Close)
    X = nifty_data[['Date_Ordinal']]  # Date as feature
    y = nifty_data['Close']  # Closing price as target

    # Create a Linear Regression model
    model = LinearRegression()
    model.fit(X, y)

    # Predict the values (fitted line)
    nifty_data['Prediction'] = model.predict(X)

    return nifty_data, model

def plot_nifty_data_with_regression(nifty_data, model):
    # Plot the actual closing prices and the linear regression line
    plt.figure(figsize=(12, 6))

    # Plot the actual closing prices
    plt.plot(nifty_data['Date'], nifty_data['Close'], label='Actual Closing Prices', color='blue', alpha=0.6)

    # Plot the linear regression prediction
    plt.plot(nifty_data['Date'], nifty_data['Prediction'], label='Linear Regression Line', color='red', linestyle='--', linewidth=2)

    # Add titles and labels
    plt.title('Nifty 50 Index - Closing Prices and Linear Regression (20 years)', fontsize=14)
    plt.xlabel('Date', fontsize=12)
    plt.ylabel('Closing Price (INR)', fontsize=12)

    # Rotate x-axis labels for better readability
    years = pd.DatetimeIndex(nifty_data['Date']).year.unique()  # Extract unique years
    plt.xticks(pd.to_datetime([f'{year}-01-01' for year in years]), rotation=45)

    # Show grid
    plt.grid(True)

    # Add legend
    plt.legend()

    # Show the plot
    plt.tight_layout()
    plt.show()

def main():
    # Fetch Nifty index data for the last 20 years
    nifty_data = fetch_nifty_data()

    # Apply Linear Regression on the data
    nifty_data, model = apply_linear_regression(nifty_data)

    # Plot the results: Actual vs Predicted (Regression Line)
    plot_nifty_data_with_regression(nifty_data, model)
    test =end_date = datetime.datetime.today().strftime('%Y-%m-%d')
    print(test)
    print(model.predict([[datetime.datetime.today().toordinal()]]))


if __name__ == "__main__":
    main()
```

```python
import yfinance as yf
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import datetime
from sklearn.linear_model import LinearRegression

# Function to fetch historical data
def fetch_index_data(ticker):
    # Calculate the start date for 20 years ago
    end_date = datetime.datetime.today().strftime('%Y-%m-%d')
    start_date = (datetime.datetime.today() - datetime.timedelta(days=20*365)).strftime('%Y-%m-%d')

    # Fetch historical data for the specified index using yfinance
    index_data = yf.download(ticker, start=start_date, end=end_date)
    print(f"{ticker} Historical Data (From {start_date} to {end_date}):")
    print(index_data.head())

    return index_data

# Function to apply linear regression on the data
# def apply_linear_regression(index_data):
#     # Prepare the data for linear regression
#     index_data['Date'] = pd.to_datetime(index_data.index)
#     index_data['Date_Ordinal'] = index_data['Date'].apply(lambda date: date.toordinal())  # Convert date to ordinal

#     # Prepare the features (Date_Ordinal) and target (Close)
#     X = index_data[['Date_Ordinal']]  # Date as feature
#     y = index_data['Close']  # Closing price as target

#     # Create a Linear Regression model
#     model = LinearRegression()
#     model.fit(X, y)

#     # Predict the values (fitted line)
#     index_data['Prediction'] = model.predict(X)

#     return index_data, model
def apply_linear_regression(index_data):
    # Prepare the data for linear regression
    index_data['Date'] = pd.to_datetime(index_data.index)
    index_data['Date_Ordinal'] = index_data['Date'].apply(lambda date: date.toordinal())  # Convert date to ordinal

    # Prepare the features (Date_Ordinal) and target (Close)
    X = index_data[['Date_Ordinal']]  # Date as feature
    y = index_data['Close']  # Closing price as target

    # Check if X is empty and handle it
    if X.empty:
        print("Error: Input data for linear regression is empty. Check the data source or ticker symbol.")
        return index_data, None  # Return None for the model

    # Create a Linear Regression model
    model = LinearRegression()
    model.fit(X, y)

    # Predict the values (fitted line)
    index_data['Prediction'] = model.predict(X)

    return index_data, model

# Function to plot the data with linear regression line
def plot_index_data_with_regression(index_data, model, title):
    # Plot the actual closing prices and the linear regression line
    plt.plot(index_data['Date'], index_data['Close'], label='Actual Closing Prices', alpha=0.6)
    plt.plot(index_data['Date'], index_data['Prediction'], label='Linear Regression Line', linestyle='--', linewidth=2)
    plt.title(title, fontsize=14)
    plt.xlabel('Date', fontsize=12)
    plt.ylabel('Closing Price (INR)', fontsize=12)
    plt.xticks(rotation=45)
    plt.grid(True)
    plt.legend()

# Main function to process and plot data for Nifty 50, Nifty Small Cap, and Nifty Mid Cap
def main():
    # Define index tickers for Nifty 50, Small Cap, and Mid Cap
    nifty50_ticker = "^NSEI"  # Nifty 50
    smallcap_ticker = "^NSESC100"  # Nifty Small Cap 100
    midcap_ticker = "^NSEMC100"  # Nifty Midcap 100

    # Fetch data for all three indices
    nifty50_data = fetch_index_data(nifty50_ticker)
    smallcap_data = fetch_index_data(smallcap_ticker)
    midcap_data = fetch_index_data(midcap_ticker)

    # Apply Linear Regression for all three indices
    nifty50_data, nifty50_model = apply_linear_regression(nifty50_data)
    smallcap_data, smallcap_model = apply_linear_regression(smallcap_data)
    midcap_data, midcap_model = apply_linear_regression(midcap_data)

    # Create a plot for all three indices
    plt.figure(figsize=(14, 8))
    print(smallcap_model)
    # Plot Nifty 50 data with linear regression line
    plt.subplot(3, 1, 1)  # First subplot (1st row)
    if nifty50_model:  # Check if model is not None
       plot_index_data_with_regression(nifty50_data, nifty50_model, 'Nifty 50 Index - Closing Prices and Linear Regression')
    else:
       plt.title('Nifty 50 Index - Data Unavailable')

    # Plot Nifty Small Cap data with linear regression line
    plt.subplot(3, 1, 2)  # Second subplot (2nd row)

    if nifty50_model:  # Check if model is not None
       plot_index_data_with_regression(smallcap_data, smallcap_model, 'Nifty 50 Index - Closing Prices and Linear Regression')
    else:
       plt.title('Midcap Index - Data Unavailable')
    # Plot Nifty Mid Cap data with linear regression line
    plt.subplot(3, 1, 3)  # Third subplot (3rd row)
    if nifty50_model:  # Check if model is not None
       plot_index_data_with_regression(midcap_data, midcap_model, 'Nifty 50 Index - Closing Prices and Linear Regression')
    else:
       plt.title('Small cap - Data Unavailable')
    # Adjust layout to prevent overlap
    plt.tight_layout()
    plt.show()

if __name__ == "__main__":
    main()
```
