```python
import yfinance as yf
import pandas as pd
from tkinter import Tk, ttk

# List of tickers
tickers = ['AAPL', 'MSFT', 'GOOGL']  # Add more tickers as needed

# Attributes to retrieve
attributes = [
    'sector',
    'sectorKey',
    'sectorDisp',
    'longBusinessSummary',
    'fullTimeEmployees',
    'trailingPE',
    'fiftyTwoWeekLow',
    'fiftyTwoWeekHigh',
    'twoHundredDayAverage',
    'currency',
    'bookValue',
    'priceToBook',
    'trailingEps'
]

# Create a DataFrame to hold the data
data = []

for ticker in tickers:
    stock = yf.Ticker(ticker)
    info = stock.info
    row = {attr: info.get(attr, 'N/A') for attr in attributes}
    row['shortName'] = info.get('shortName', 'N/A')  # Add short name to row
    row['Ticker'] = ticker  # Add ticker to row
    data.append(row)

# Convert the list of dictionaries to a DataFrame
df = pd.DataFrame(data)

# Select only relevant columns for display
df_display = df[['shortName', 'Ticker'] + attributes]

# Create the main application window
root = Tk()
root.title("Stock Data")

# Create a Treeview widget
tree = ttk.Treeview(root)

# Define columns
tree['columns'] = list(df_display.columns)
tree['show'] = 'headings'  # Hide the first empty column

# Create headings
for column in tree['columns']:
    tree.heading(column, text=column)

# Add data to the treeview
for index, row in df_display.iterrows():
    tree.insert("", "end", values=list(row))

# Pack the treeview into the window
tree.pack(fill="both", expand=True)

# Run the application
root.mainloop()
```
