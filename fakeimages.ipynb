```python
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
import tensorflow_hub as hub

# Install tensorflow_hub library (if not already installed)
!pip install tensorflow_hub

# Load the pre-trained PGGAN model from tensorflow_hub
generator = hub.load("https://tfhub.dev/google/progan-128/1").signatures['default']

# Function to generate beautiful human faces
def generate_faces(num_faces=5):
    noise = tf.random.normal([num_faces, 512])  # Generate random noise as input to the generator

    # Generate faces using the pre-trained PGGAN model
    generated_images = generator(noise)['default']
    generated_images = (generated_images.numpy() + 1) / 2.0  # Rescale to [0, 1]

    # Display the generated images
    plt.figure(figsize=(num_faces, 1))
    for i in range(num_faces):
        plt.subplot(1, num_faces, i+1)
        plt.imshow(generated_images[i])
        plt.axis('off')
    plt.show()

# Generate beautiful human faces
generate_faces(num_faces=5)
```
