```python
from transformers import MBartForConditionalGeneration, MBart50TokenizerFast

with open("/content/transformer_detail.txt", 'r') as file:
    file_content = file.read()

# Print the content of the file
print("File Content:\n", file_content)
article_english = file_content
model = MBartForConditionalGeneration.from_pretrained("facebook/mbart-large-50-many-to-many-mmt")
tokenizer = MBart50TokenizerFast.from_pretrained("facebook/mbart-large-50-many-to-many-mmt")
tokenizer.src_lang = "en_XX"
# translate English to Tamil

encoded_hi = tokenizer(article_english, return_tensors="pt")
generated_tokens = model.generate(
    **encoded_hi,
    forced_bos_token_id=tokenizer.lang_code_to_id["ta_IN"]
)
response = tokenizer.batch_decode(generated_tokens, skip_special_tokens=True)
print(response)
```

```python
# Use a pipeline as a high-level helper
from transformers import pipeline

messages = [
    {"role": "user", "content": "Who are you?"},
]
pipe = pipeline("text-generation", model="microsoft/Phi-3.5-mini-instruct", trust_remote_code=True)
pipe(messages)
```
